```python
import numpy as np
import matplotlib.pyplot as plt

# Time parameters
t_start = 0
t_end = 10
dt = 0.01
time = np.arange(t_start, t_end, dt)

# Wave parameters
freq1 = 1.0  # Frequency of the first wave (Hz)
freq2 = 1.5  # Frequency of the second wave (Hz)
phase_diff = np.linspace(0, 2*np.pi, len(time))  # Gradually changing phase difference

# Generate sine waves
wave1 = np.sin(2*np.pi*freq1*time)
wave2 = np.sin(2*np.pi*freq2*time + phase_diff)

# Add spikes for synchronization
spike_indices = [int(len(time) * i / len(phase_diff)) for i in range(len(phase_diff))]
wave1[spike_indices] = 2  # Add spikes to the first wave
wave2[spike_indices] = 2  # Add spikes to the second wave

# Plot waves
plt.figure(figsize=(10, 6))
plt.plot(time, wave1, label='Wave 1')
#plt.plot(time, wave2, label='Wave 2')
plt.title('Synchronized Waves with Spikes')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
